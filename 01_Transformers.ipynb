```python
import pandas as pd
from transformers import pipeline

# Defining the text variable within the current scope
text = """Dear Amazon, last week I ordered an Optimus Prime action figure \
from your online store in Germany. Unfortunately, when I opened the
package, \
I discovered to my horror that I had been sent an action figure of Megatron
\
instead! As a lifelong enemy of the Decepticons, I hope you can understand
my \
dilemma. To resolve the issue, I demand an exchange of Megatron for the \
Optimus Prime figure I ordered. Enclosed are copies of my records
concerning \
this purchase. I expect to hear from you soon. Sincerely, Bumblebee."""

# Initialize the sentiment analysis pipeline
classifier = pipeline("sentiment-analysis")

# Perform sentiment analysis and convert the results to a DataFrame
outputs = classifier(text)
pd.DataFrame.from_records(outputs)
```

# Named Entities Recognition (NER)

```python
ner_tagger = pipeline("ner", aggregation_strategy="simple")
outputs = ner_tagger(text)
pd.DataFrame.from_records(outputs)
```

# Question Answering


```python
reader = pipeline("question-answering")
question = "What does the customer want?"
outputs = reader(question=question, context=text)
pd.DataFrame.from_records([outputs])
```

# Summarization

```python
summarizer = pipeline("summarization")
outputs = summarizer(text, max_length=45,
clean_up_tokenization_spaces=True)
print(outputs[0]['summary_text'])
```

# Reply to the Custome

```python
from transformers import set_seed
from transformers import set_seed, pipeline
set_seed(42)
generator = pipeline("text-generation")
response = "Dear Bumblebee, I am sorry to hear that your order was mixed up."
prompt = text + "\n\nCustomer service response:\n" + response
outputs = generator(prompt, max_length=200)
print(outputs[0]['generated_text'])
```

# Tokenization

```python
text = "Tokenizing text is a core task of NLP."
tokenized_text = list(text)
print(tokenized_text)
```

# Simple Python-based text Tokenization
The function list(text) doesn’t apply any tokenization strategy in the NLP sense. Instead, the list() function is a Python built-in function that converts any iterable (in this case, the string text) into a list of its elements.

Since a string is an iterable of characters, calling list(text) breaks the string into its individual characters, not words or subwords.

------------------
# Why does it default to character tokenization?
Strings in Python are iterable, and by default, iterating over a string results in character-by-character iteration.

So, when you pass a string to list(), it interprets the string as a sequence of individual characters, and each character becomes a separate element in the resulting list.

To explicitly tokenize based on words, you would need to either use the split() method, as mentioned earlier, or use an NLP-specific library like nltk or spaCy for more advanced tokenization strategies.

------------------------------
#Summary
The system doesn't "know" you're asking for character tokenization.

It's just the default behavior of Python's list() function when applied to a string, as Python treats strings as sequences of characters.

If you need word or subword tokenization (which is more typical in NLP), you would need to use a specific method like split() or a dedicated tokenizer from an NLP library.


```python
token2idx = {}
for idx, unique_char in enumerate(set(tokenized_text)):
token2idx[unique_char] = idx
print(token2idx)
```
